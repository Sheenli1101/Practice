```python
import csv
```

```python
reader = csv.reader(open('金鐘獎101_106-2.csv', 'r', encoding='utf_8_sig'))
```

```python
data = []
for i in reader:
    data.append(i)
```

```python
data
```

```python
columns = ['年度', '類別', '獎項', '得獎名單']
```

```python
col = columns.index('類別')
```

```python
for e in data:
    print(e[col])
```

```python
gaol = {}
for e in data:
    if e[col] in gaol:
        gaol[e[col]] += 1
    else:
        gaol[e[col]] = 1
```

```python
print(gaol)
```
