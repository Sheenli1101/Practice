# 傳遞參數兩種模式(C語言)
- pass by value (只有值進去)
- pass by reference (本尊直接跳進去函式)

# Python的傳遞模式為 Pass by Object reference

```python
def f(y):
    y.append(1)

x = [0]
f(x)
print(x)

# pass by value 應為 [0]
# pass by reference 應為 [0, 1]
```

```python
def ff(y):
    y += 1


x = 0
ff(x)
print(x)
```
