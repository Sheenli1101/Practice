# How do I filter rows of a pandas DataFrame by column value?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
boolean = []
for length in movies.duration:
    if length >= 200:
        boolean.append(True)
    else:
        boolean.append(False)
        
```

```python
len(boolean)
```

```python
is_long = pd.Series(boolean)
```

```python
is_long.head()
```

```python
# pandas很聰明，用Series去比較一個數字，就知道要用Series的各元素去跟一個數字比
is_long = movies.duration >= 200
is_long.head()
```

```python
#[]裡面是要創造布林值
movies[movies.duration <= 200]
movies[movies.duration >= 150]
```

```python
movies[is_long]
```

```python
# 雙重條件 use & 
movies[(movies.duration <= 200) & (movies.duration>=150)]['genre']
```

```python
movies.loc[movies.duration <= 200,['genre','content_rating']]
```
