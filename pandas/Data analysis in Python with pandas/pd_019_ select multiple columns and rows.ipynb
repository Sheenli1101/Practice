# How do I select multiple rows and columns from Dataframe?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
# loc[列,欄] 以單一列作選擇會變成Series
ufo.loc[0,:]
```

```python
# 選擇index = 0, 1, 2的資料，欄位全選
ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[0:2, :]
```

```python
# 以單一欄位做選擇，會形成Series
ufo.loc[: , 'City']
```

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.loc[0:2, 'City':'State']
```

```python
ufo.head(3).drop('Time', axis = 1)
```

```python
# 沒辦法設定欄位範圍
ufo[ufo.City == 'Oakland']
```

```python
# 更有彈性的作法
ufo.loc[(ufo.City == 'Willingboro') | (ufo.City == 'Oakland') ,'City':'Colors Reported']
```

```python
# iloc 跟range相同，不包括上限
# loc 上下限皆包括
# iloc 欄位用數字代表
ufo.iloc[:, 0:4]
```

```python
ufo.iloc[0:3, :]
```

```python
# = ufo[['City', 'State']]
ufo.loc[:, ['City', 'State']]
```

```python
ufo.iloc[0:2,:]
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col = 'country')
```

```python
drinks.head()
```

```python
drinks.loc['Afghanistan':'Andorra':,['beer_servings','wine_servings']]
```

```python
drinks.loc['Afghanistan':'Andorra':,'beer_servings':'wine_servings']
```

```python
# index 不再只是數字...
# 能用loc 跟iloc就不用ix!
drinks.ix['Albania',0]
```

```python
drinks.ix[1, 'beer_servings']
```

```python
drinks.ix['Albania':'Andorra', 0:2]
```

```python
drinks.ix['Albania':'Andorra', ['beer_servings','spirit_servings']]
```
